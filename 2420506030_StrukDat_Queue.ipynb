```python
#Implementasi Queue menggunakan list
class QueueList:
    def __init__(self):
        self.queue = []

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop(0)
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if  not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

#Contoh penggunaan

queue = QueueList()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)

print("Queue setelah enqueue:", queue.queue)
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

```python
#Implementasi Queue dengan collections.deque

from collections import deque

class QueueDeque:
    def __init__(self):
        self.queue = deque()

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        return self.queue.popleft() if not self.is_empty() else None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

queue = QueueDeque()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)

print("Queue setelah enqueue:", list (queue.queue))
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

```python
#Implementasi Queue dengan Linked List

class Node:
      def __init__(self, data):
          self.data = data
          self.next = None

class QueueLinkedList:
    def __init__(self):
        self.front = self.rear = None

    def enqueue(self, data):
        new_node = Node(data)
        if self.rear is None:
            self.front = self.rear = new_node
            return
        self.rear.next = new_node
        self.rear = new_node

    def dequeue(self):
        if self.front is None:
            return None
        dequeued_element = self.front.data
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        return dequeued_element

    def is_empty(self):
        return self.front is None

    def front_element(self):
        return self.front.data if self.front else None

    def rear_element(self):
        return self.rear.data if self.rear else None

queue = QueueLinkedList()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front_element())
print("Rear element:", queue.rear_element())
```

```python
#Latihan 1
class AntrianBioskop:
    def __init__(self, kapasitas_maks=10):
        self.antrian = []
        self.kapasitas_maks = kapasitas_maks

    def kode_pelanggan(self, nama, npm):
        return nama[:2].upper() + npm[-2:]

    def tambah_pelanggan(self, kode_pelanggan):
        if '7' in kode_pelanggan:
            print(f"Pelanggan VIP {kode_pelanggan}, Anda langsung mendapat kursi!")
        elif len(self.antrian) < self.kapasitas_maks:
            self.antrian.append(kode_pelanggan)
            print(f"Pelanggan {kode_pelanggan}, Anda telah masuk antrian.")
        else:
            print(f"Antrian penuh! {kode_pelanggan}, Anda harus menunggu.")

    def melayani_pelanggan(self):
        if not self.kosong():
            pelanggan_dilayani = self.antrian.pop(0)
            print(f"Pelanggan {pelanggan_dilayani} telah dilayani.")
            return pelanggan_dilayani
        else:
            print("Antrian kosong!")
            return None

    def pelanggan_pertama(self):
        return self.antrian[0] if not self.kosong() else None

    def pelanggan_terakhir(self):
        return self.antrian[-1] if not self.kosong() else None

    def kosong(self):
        return len(self.antrian) == 0

    def jumlah_pelanggan(self):
        return len(self.antrian)


bioskop = AntrianBioskop()

while True:
    print("\n=== Sistem Antrian Bioskop ===")
    print("1. Tambah Pelanggan")
    print("2. Layani Pelanggan")
    print("3. Lihat Pelanggan Pertama")
    print("4. Lihat Pelanggan Terakhir")
    print("5. Lihat Jumlah Pelanggan")
    print("6. Keluar")

    pilihan = input("Pilih menu (1-6): ")

    if pilihan == "1":
        nama = input("Masukkan nama pelanggan: ")
        npm = input("Masukkan NPM pelanggan: ")
        kode_pelanggan = bioskop.kode_pelanggan(nama, npm)
        bioskop.tambah_pelanggan(kode_pelanggan)

    elif pilihan == "2":
        bioskop.melayani_pelanggan()

    elif pilihan == "3":
        pelanggan_pertama = bioskop.pelanggan_pertama()
        if pelanggan_pertama:
            print(f"Pelanggan pertama: {pelanggan_pertama}")
        else:
            print("Antrian kosong.")

    elif pilihan == "4":
        pelanggan_terakhir = bioskop.pelanggan_terakhir()
        if pelanggan_terakhir:
            print(f"Pelanggan terakhir: {pelanggan_terakhir}")
        else:
            print("Antrian kosong.")

    elif pilihan == "5":
        print(f"Total pelanggan dalam antrian: {bioskop.jumlah_pelanggan()}")

    elif pilihan == "6":
        print("Keluar dari sistem.")
        break

    else:
        print("Pilihan tidak valid. Silakan pilih menu yang tersedia.")


```

```python
#Latihan 2
class Fotokopi:
    def __init__(self):
        self.antrian = []

    def id_pesanan(self,nama,npm):
        return nama[:3] + npm[-2:]

    def enqueue(self, order_id, num_pages):
        if num_pages < 10:
            print(f"Pesanan {order_id} sebanyak {num_pages} halaman masuk ke fast track, pesanan akan langsung diproses")
        else:
            self.antrian.append((order_id, num_pages))
            print(f"Pesanan {order_id} sebanyak {num_pages} halaman masuk ke antrian")

            if num_pages > 100:
                self.antrian.append((order_id, "tahap 2 "))
                print(f"Pesanan {order_id} lebih dari 100 halaman, pesanan akan dikerjakan dalam dua tahap")

    def dequeue(self):
        if not self.is_empty():
            order_id, num_pages = self.antrian.pop(0)
            print(f"Melayani pesanan({order_id})({num_pages}) halaman")
            return order_id
        else:
            print("Antrian kosong!")
            return None

    def front(self):
        return self.antrian[0] if not  self.is_empty() else None

    def rear(self):
        return self.antrian[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.antrian) == 0

    def size(self):
        return len(self.antrian)

fotokopi = Fotokopi()

while True:
    print("\n=== Sistem Antrian Fotokopi ===")
    print("1. Tambah Pesanan")
    print("2. Layani Pesanan")
    print("3. Lihat Pesanan Pertama")
    print("4. Lihat Pesanan Terakhir")
    print("5. Lihat Jumlah Pesanan")
    print("6. Keluar")

    pilihan = input("Pilih menu (1-6): ")

    if pilihan == "1":
        nama = input("Masukkan Nama : ")
        npm = input("Masukkan NPM : ")
        num_pages = int(input("Masukkan jumlah halaman: "))

        order_id = fotokopi.id_pesanan(nama, npm)
        fotokopi.enqueue(order_id, num_pages)

    elif pilihan == "2":
        fotokopi.dequeue()

    elif pilihan == "3":
        print(f"Pesanan pertama: {fotokopi.front()}")

    elif pilihan == "4":
        print(f"Pesanan terakhir: {fotokopi.rear()}")

    elif pilihan == "5":
        print(f"Total pesanan dalam antrian: {fotokopi.size()}")

    elif pilihan == "6":
        print("Keluar dari sistem.")
        break

    else:
        print("Pilihan tidak valid. Silakan pilih menu yang tersedia.")
```

```python
#Latihan 3
class Parkiran:
    def __init__(self, kapasitas=15):
        self.antrian = []
        self.kapasitas = kapasitas

    def id_parkir(self,nama_kota,npm):
        return nama_kota[:3] + npm[-2:]

    def enqueue(self,vehicle_number, jenis_mobil):
        if jenis_mobil == 1:
            print(f"Antrian {vehicle_number} dengan jenis mobil {jenis_mobil} dapat masuk ke parkiran terlebih dahulu")
        elif len(self.antrian) < self.kapasitas:
            self.antrian.append((vehicle_number, jenis_mobil))
            print(f"Antrian {vehicle_number} dengan jenis mobil  {jenis_mobil} harus menunggu parkiran")
        else:
            print(f"Antrian {vehicle_number} melebihi batas parkir, harap tunggu")

    def dequeue(self):
        if not self.is_empty():
            vehicle_number, jenis_mobil = self.antrian.pop(0)
            print(f"Melayani pesanan {vehicle_number}){jenis_mobil} halaman")
            return vehicle_number
        else:
            print("Antrian kosong!")
            return None

    def front(self):
        return self.antrian[0] if not  self.is_empty() else None

    def rear(self):
        return self.antrian[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.antrian) == 0

    def size(self):
        return len(self.antrian)

parkiran = Parkiran()

while True:
    print("\n=== Sistem Antrian Parkir ===")
    print("1. Tambah Antrian")
    print("2. Layani Antrian")
    print("3. Lihat Antrian Pertama")
    print("4. Lihat Antrian Terakhir")
    print("5. Lihat Jumlah Antrian")
    print("6. Keluar")

    pilihan = input("Pilih menu (1-6): ")

    if pilihan == "1":
        nama_kota = input("Masukkan Nama kota asal anda : ")
        npm = input("Masukkan NPM : ")
        jenis_mobil = int(input("Masukkan jenis mobil: (kode 1 untuk listrik dan 2 untuk bensin) "))

        vehicle_number = parkiran.id_parkir(nama_kota, npm)
        parkiran.enqueue(vehicle_number, jenis_mobil)

    elif pilihan == "2":
        parkiran.dequeue()

    elif pilihan == "3":
        print(f"Antrian pertama: {parkiran.front()}")

    elif pilihan == "4":
        print(f"Antrian terakhir: {parkiran.rear()}")

    elif pilihan == "5":
        print(f"Total Antrian dalam parkiran: {parkiran.size()}")

    elif pilihan == "6":
        print("Keluar dari sistem.")
        break

    else:
        print("Pilihan tidak valid. Silakan pilih menu yang tersedia.")
```

```python
#latihan 4
class AntrianSertifikat:
    def __init__(self):
        self.antrian = []
        self.antrian_reguler = []
        self.antrian_berprestasi = []

    def id_antrian(self, program_studi, npm):
        return program_studi[0].upper() + npm[-2:]

    def enqueue(self, student_id, berprestasi=False):
        if berprestasi:
            print(f"Mahasiswa {student_id} (berprestasi) dapat langsung mengambil sertifikat.")
        elif len(self.antrian) < 20:
            self.antrian.append(student_id)
            print(f"Mahasiswa {student_id} telah ditambahkan ke dalam antrian.")
        else:
            if len(self.antrian_reguler) + len(self.antrian_berprestasi) == 0:
                self.antrian_reguler = self.antrian[:]
                self.antrian.clear()
            self.antrian_reguler.append(student_id)
            print(f"Mahasiswa {student_id} masuk ke jalur reguler.")

    def dequeue(self):
        if self.antrian:
            student_id = self.antrian.pop(0)
            print(f"Mahasiswa {student_id} telah dilayani.")
            return student_id
        elif self.antrian_berprestasi:
            student_id = self.antrian_berprestasi.pop(0)
            print(f"Mahasiswa {student_id} telah dilayani dari jalur berprestasi.")
            return student_id
        elif self.antrian_reguler:
            student_id = self.antrian_reguler.pop(0)
            print(f"Mahasiswa {student_id} telah dilayani dari jalur reguler.")
            return student_id
        else:
            print("Antrian kosong!")
            return None

    def front(self):
        return self.antrian[0] if self.antrian else (self.antrian_berprestasi[0] if self.antrian_berprestasi else (self.antrian_reguler[0] if self.antrian_reguler else None))

    def rear(self):
        return self.antrian[-1] if self.antrian else (self.antrian_reguler[-1] if self.antrian_reguler else (self.antrian_berprestasi[-1] if self.antrian_berprestasi else None))

    def is_empty(self):
        return len(self.antrian) == 0 and len(self.antrian_reguler) == 0 and len(self.antrian_berprestasi) == 0

    def size(self):
        return len(self.antrian) + len(self.antrian_reguler) + len(self.antrian_berprestasi)

antrian_sertifikat = AntrianSertifikat()

while True:
    print("\n=== Sistem Antrian Sertifikat ===")
    print("1. Tambah Antrian")
    print("2. Layani Antrian")
    print("3. Lihat Antrian Pertama")
    print("4. Lihat Antrian Terakhir")
    print("5. Lihat Jumlah Antrian")
    print("6. Keluar")

    pilihan = input("Pilih menu (1-6): ")

    if pilihan == "1":
        program_studi = input("Masukkan Program Studi: ")
        npm = input("Masukkan NPM: ")
        prestasi = input("Apakah mahasiswa pernah mengikuti kompetisi tingkat nasional? (y/n): ").strip().lower()
        student_id = antrian_sertifikat.id_antrian(program_studi, npm)
        antrian_sertifikat.enqueue(student_id, berprestasi=(prestasi == 'y'))

    elif pilihan == "2":
        antrian_sertifikat.dequeue()

    elif pilihan == "3":
        print(f"Antrian pertama: {antrian_sertifikat.front()}")

    elif pilihan == "4":
        print(f"Antrian terakhir: {antrian_sertifikat.rear()}")

    elif pilihan == "5":
        print(f"Total Antrian: {antrian_sertifikat.size()}")

    elif pilihan == "6":
        print("Keluar dari sistem.")
        break

    else:
        print("Pilihan tidak valid. Silakan pilih menu yang tersedia.")

```
